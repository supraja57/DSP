```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Loading the dataset from sklearn library
from sklearn.datasets import load_boston
boston = load_boston()
```

## Creating a dataframe

```python
boston_df = pd.DataFrame(data = boston.data,columns = boston.feature_names)
boston_df
```

```python
boston_df['MEDV'] = boston.target
boston_df.head()
```

```python
# split the data into training and testing; Build the model on training data an
```

```python
# split the data into train and test
x = boston_df.drop(['MEDV'],axis=1)
y = boston_df.MEDV
```

```python

# divide the x and y into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

## Checking the intercept and coefficient

```python
coefficients = pd.DataFrame([x_train.columns,lm.coef_])
coefficients
```

```python
coefficients = pd.DataFrame([x_train.columns,lm.coef_]).T
coefficients
```

## Model Prediction training data

```python
y_train_pred = lm.predict(x_train)
y_train_pred
```

## R2 score

```python
from sklearn.metrics import r2_score
print('R^2:',r2_score(y_train,y_train_pred))
```

```python
from sklearn.metrics import r2_score
print('R^2:',r2_score(y_train,y_train_pred))
print('Adjusted R^2:',1-(1-r2_score(y_train,y_train_pred))*(len(x_train)-1)/
      (len(x_train)-x_train.shape[1]-1))
```

```python

plt.scatter(y_train,y_train_pred)
plt.xlabel('price')
plt.ylabel('Model Predicted Price')
```

## Residual Plot

```python

plt.scatter(y_train_pred,y_train-y_train_pred)
plt.xlabel("Predicted Values")
plt.ylabel("residuals")
```

## Predicting on Test Data

```python
y_test_pred = lm.predict(x_test)
y_test_pred
```

```python
from sklearn.metrics import r2_score
print('R^2:',r2_score(y_test,y_test_pred))
print('Adjusted R^2:',1-(1-r2_score(y_test,y_test_pred))*(len(x_test)-1)/
      (len(x_test)-x_test.shape[1]-1))
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print('R^2:', r2_score(y_test,y_test_pred))


print('Adjusted R^2:', 1- (1-r2_score(y_test, y_test_pred))*(len(x_test)-1)/
                                  (len(x_test)-x_test.shape[1]-1))


print('MAE:', mean_absolute_error(y_test, y_test_pred))

print('MSE:', mean_squared_error(y_test, y_test_pred))

print('RMSE', np.sqrt(mean_squared_error(y_test, y_test_pred)))
```
